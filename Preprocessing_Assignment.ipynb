**## Handling missing values of cloumns Cabin and Embarked ##**

**##Hnadling missing values of coulmn 'Age' . Filling with Median value ##**

```python
sel_cols = df[['Age']]
sel_cols.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer = imputer.fit(sel_cols)
sel_cols = imputer.transform(sel_cols)
```

```python
type(sel_cols)
```

```python
sel_cols = pd.DataFrame(sel_cols, columns=['Age'])
type(sel_cols)
```

```python
sel_cols.isna().sum()
```

```python
df = df.drop(['Age'],axis=1)
df.head()
```

```python
sel_cols.shape
```

```python
sel_cols.head()
```

**##Concatenting Age column with rest of the columns after setting the index of Column Age to PassengerID##**

```python
df = pd.concat([df,sel_cols.set_index(df.index)], axis = 1)
df.head()
```

##Handling Outliers##
1. Age
2. Fare
3. SibSp  

```python
plt.boxplot(df['Age'])
plt.show()
```

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
print(iqr)
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(lower_bound)
print(upper_bound)
```

```python
outlier = []
for i in df['Age']:
  if i < lower_bound or i > upper_bound:
    outlier.append(i)
```

```python
outlier
```

```python
ind = df[df['Age'] > upper_bound].index
df.drop(ind, inplace=True)
```
